### Operación entre matrices

```python
# Suma Mat 1 + Mat 2

mat1+mat2
```

La operación entre matrices de diferente forma y dimensión diferentea 1 no es permitida en Numpy

```python
# Suma Mat 1 + Mat 3

mat4 = mat1+mat3
mat4
```

Para realizar la operación, la matriz de dimensión 1 se expande las dimensiones necesarias restantes en base a sí misma.

# Ejercicio 5

Extracción de primera 

```python
extraccion1 = mat4[1, 0]
extraccion2 = mat4[2, 2]

print('1ra extracción. Segunda fila, primera columna: ', extraccion1)
print('2da extracción. Tercera fila, tercera columna: ', extraccion2)

print('La suma entre extracciones es: ', extraccion1+extraccion2)
```
